```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
#create set of lat. and long combos
lat=np.random.uniform(low=-90.000, high=90.000, size=2000)
long=np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_long=zip(lat,long)
lat_long
```

```python
coordinates=list(lat_long)
```

```python
from citipy import citipy as cp
```

```python
# make a list
cities=[]

# ID nearest city for lat and long combo
for coordinate in coordinates:
    city=cp.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # if the city is unique, add to cities list
    if city not in cities:
        cities.append(city)

# print
len(cities)
```

```python
import requests
from config import weather_api_key
```

```python
#starting url for 
url="http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
# make an empty list
city_data=[]

#print the beginning of the log
print("Beginning Data Retrieval       ")
print("-------------------------------")

#make counters
record_count=1
set_count=1

#loop through cities in list
for i, city in enumerate(cities):
    if(i %50 ==0 and i >=50):
        set_count +=1
        record_count=1

    city_url = url + "&q=" + city
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    record_count+=1

 #run an API request for each city
    try:
        #parse JSON and retrieve data
        city_weather=requests.get(city_url).json()
        #parse out needed data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        #convert date to standard
        city_date=datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        #add city info to city_data list
        city_data.append({"City": city.title(),
                              "Latitude": city_lat,
                              "Longitude": city_lng,
                              "Max Temp": city_max_temp,
                              "Humidity": city_humidity,
                              "Cloudiness": city_clouds,
                              "Wind Speed": city_wind,
                              "Country": city_country,
                              "Date": city_date})
#if an error occurs, skip the city
    except:
        print("City not found. Skipping...")
        pass
# indicate when data loading is complete
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```

```python
#convert array of dictionaries to pandas df
city_data_df=pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
new_column_order=["City","Country","Date","Latitude","Longitude","Max Temp","Humidity","Cloudiness","Wind Speed"]
city_data_df=city_data_df[new_column_order]
city_data_df.head(10)
```

```python
csv_data_file="weather_database.csv"
city_data_df.to_csv(csv_data_file, index_label="City ID")
```
